```python
from datetime import date, timedelta,datetime

import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
import lightgbm as lgb
```

```python
pred = pd.read_csv("aaaa.csv")
sub = pd.read_csv("holiday_substrace.csv")
```

```python
del(sub["Unnamed: 0"])
sub["diffrence"] = sub["unit_sales"]
del(sub["unit_sales"])
```

```python
del(pred["Unnamed: 0"])
pred["date"] = pred["level_2"]
del(pred["level_2"])
```

```python
df_test = pd.read_csv(
    "test.csv", usecols=[0, 1, 2, 3],
      # , date_parser=parser
)
```

```python
df_test.dtypes
```

```python
pred["diffrence"] = pred["diffrence"].apply(lambda x : -x)
```

```python
pred = pd.merge(pred,sub,how="outer",on=["store_nbr","item_nbr"])
pred["unit_sales"] = pred["unit_sales"].apply(lambda x:x if 0<x else 0)
```

```python
pred[pred["date"] == "2017-08-24"]["unit_sales"] += pred[pred["date"] == "2017-08-24"]["diffrence"]
```

```python
pred.set_index(["store_nbr","item_nbr","date"],inplace=True)
```

```python
submission = df_test[["id"]].join(pred, how="left").fillna(0)
```

```python
submission["unit_sales"] = np.clip(np.expm1(submission["unit_sales"]), 0, 1000)
submission.to_csv('lgb6.csv', float_format='%.4f', index=None)
```
